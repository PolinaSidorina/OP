Коза, привязанная за веревку в к ограде, пасется на круглом поле. Какова должна быть длина веревки, чтобы у козы был доступ ровно к половине площади всего поля? Радиус поля считается равным $1$.

Покажите, что задача сводится к решению уравнения
$$
\sin\alpha - \alpha\cos\alpha = \frac{\pi}{2}, \qquad (1)
$$
откуда можно найти длину веревки
$$
r = 2\cos\frac{\alpha}{2}.
$$

# Решение:
В первую очередь я немного перерисовала картинку, чтобы мне было удобней ориентироваться:

Я воспользуюсь численным подходом к решению. 

Пусть загон представляет собой единичный круг площадью $\pi$, с центром в точке O, коза привязана к загону в точке T. 

Площадь выпаса козы выделена на рисунке голубым цветом и равна $\frac{\pi}{2}$. 
Моя задача - найти $r$.

Составим несколько уравнений по получившемуся рисунку:

Так как  ΔPTO-равнобедренный треугольник, то
$$
\alpha = \pi - 2θ, \qquad (1)
$$
По теореме синусов в треугольнике ΔPTO
$$
\sinθ= \frac{\sin\alpha}{r}, \qquad (2)   
$$
По теореме косинусов в треугольнике ΔPTO
$$
r^2=1+1-2\cos\alpha=2(1-\cos\alpha), \qquad (3)
$$

$$
1=1+r^2-2r\cosθ => θ=\arccos\frac{r}{2}, \qquad (4)
$$

Площадь кругового сектора TPK:
$$
S_{TPK}=\pi r^2 \frac{θ}{2\pi}=\frac{r^2 θ}{2},
$$
площадь кругового сектора OTP:
$$
S_{OTP}=\pi \frac{\alpha}{2\pi}=\frac{\alpha}{2},
$$
площадь тругольника ΔOTP:
$$
S_{ΔOTP}=\frac{1}{2}r \sinθ,
$$
Используя $\qquad (2)$
$$
S_{ΔOTP}=\frac{1}{2} \sin\alpha\
$$
Таким образом, площадь, по которой должна ходить коза, равна
$$
S=2(S_{TPK}+S_{OTP}-S_{ΔOTP})=\frac{\pi}{2}
$$
Тогда
$$
S=r^2 θ+\alpha-\sin\alpha=\frac{\pi}{2}
$$
Используя $\qquad (1)$
$$
θ=\frac{1}{2}(\pi-\alpha)
$$
$$
\sin\alpha+(\pi-\alpha)\cos\alpha=\frac{\pi}{2}
$$
Таким образом задача сводится к решению уравнения
$$
\sin\alpha - \alpha\cos\alpha = \frac{\pi}{2}, \qquad (5)
$$
откуда можно найти длину веревки
$$
r = 2\cos\frac{\alpha}{2}.
$$


Уравнение (5) можно решить с помощью алгоритма поиска корня Ньютона-Рафсона, реализованного в Python как scipy.optimize.newton:

```python
import numpy as np
from scipy.optimize import newton

def func(alpha):
    return np.sin(alpha) + (np.pi - alpha) * np.cos(alpha) - np.pi / 2

theta_guess = np.pi / 2
alpha = newton(func, theta_guess)
print(alpha)
```

Это значение для $\alpha$ находится в радианах. 

Соответствующая длина каната $r$ выводится из (3) в виде:

```python
r = np.sqrt(2 * (1-np.cos(alpha)))
print(r)
```

Изменение пастбищной части загона $А/\pi$ в зависимости от длины каната $r$
 можно изобразить следующим образом:




```python
import matplotlib.pyplot as plt

rgrid = np.linspace(0, 2, 100)
cos_alpha = 1 - rgrid**2 / 2
alpha = np.arccos(cos_alpha)
theta = (np.pi - alpha) / 2
A = rgrid**2 * theta + alpha - np.sin(alpha)

plt.plot(rgrid, A / np.pi, color='green')
plt.hlines(0.5, 0, r, colors='grey', linestyles='--')
plt.vlines(r, 0, 0.5, colors='grey', linestyles='--')
plt.xlim(0, 2)
plt.ylim(0, 1)
plt.yticks([0, 0.5, 1])
plt.xlabel('Длина веревки')
plt.ylabel('Пастбищная часть загона')
plt.show()
```
